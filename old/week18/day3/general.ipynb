```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x ** 2

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('My first plot')
plt.annotate('best value', xy=(2, 2.5), xytext=(1, 30),arrowprops=dict(facecolor='black', shrink=0.05),
            )
plt.plot(x, y,color='red',linestyle='dashed',linewidth=2,marker='o',markerfacecolor='blue',markersize=2)
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.tan(x)

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('My first plot')

plt.plot(x, y,color='red',linestyle='dashed',linewidth=2,marker='o',markerfacecolor='blue',markersize=2)
plt.show()
```

```python
#plot scatter of random numbers

figure,axes=plt.subplots()

x=np.random.rand(10)*100
y=np.random.rand(10)*10


plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('My second plot')

axes.legend('scatter plot',loc='upper right')

axes.scatter(x,y,label='scatter plot')
plt.show()

figure.savefig('scatter_plot.png')
```

```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = x ** 2
y2=x**3

fig,axes=plt.subplots()

plt.xlabel('x axis')
plt.ylabel('y axis')

plt.title('My first plot')

    
axes.plot(x, y,color='red',linestyle='dashed',linewidth=2,marker='o',markerfacecolor='blue',markersize=2,label='line1')
axes.plot(x, y2,color='green',linestyle='dashed',linewidth=2,marker='o',markerfacecolor='blue',markersize=2,label='line2')
axes.legend(loc='upper right')
plt.show()
```
