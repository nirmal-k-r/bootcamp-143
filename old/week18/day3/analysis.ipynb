```python
# Plotting directly from the DataFrame
df.plot(kind='scatter',x='Rating',y='Unit price',color='red')
# df.plot(kind='scatter',x='Rating',y='Unit price',color='#2d4acc')
plt.show()
```

```python
sales_by_gender=df.groupby('Gender').count()['Invoice ID']
sales_by_gender
```

```python
sales_by_city=df.groupby('City').count()['Invoice ID']
sales_by_city
```

```python
#use matplotlib to plot the data
import matplotlib.pyplot as plt

# Create a figure and axis
figure, axes = plt.subplots()

figure.set_size_inches(8,5)
axes.set_xlabel('City')
axes.set_ylabel('Total Sales')
axes.set_title('Analysis of sales by city')


# Plot the data
axes.barh(y=sales_by_city_total.index,width=sales_by_city_total['Total'],color='red')

plt.show()
```
